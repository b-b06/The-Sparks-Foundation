```python
ds=pd.read_csv('Iris.csv')
ds
```

```python
ds.drop('Id',inplace=True,axis=1)
sns.pairplot(ds,hue='Species')
```

```python
x=ds.iloc[:,:-1].values
y=ds['Species']
```

### Step5: Label Encoding 

```python
le=LabelEncoder()
y=le.fit_transform(y)
y
```

### Step6:  Data Evaluation

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.36)
```

```python
DT=DecisionTreeClassifier()
DT.fit(x_train,y_train)
```

```python
print("{:.2f}".format(DT.score(x_train,y_train)))
print("{:.2f}".format(DT.score(x_test,y_test)))
```

```python
pred=DT.predict(x_test)
d=pd.DataFrame({'Original Data':y_test,'Predicted Data':pred})
d
```

```python
print(classification_report(y_test, pred))
```
